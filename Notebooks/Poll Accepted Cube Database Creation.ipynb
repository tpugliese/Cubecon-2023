```python
# Split the URL for the ID
poll_cubes['ID']=poll_cubes['URL'].str.split('/',expand=True).loc[:,5]
# Remove unnessecary columns
poll_cubes_slice = poll_cubes.loc[:,['Cube Name', 'ID', 'Poll Winner']]

# Placeholder for replacing IDs before URL gets created
# poll_cubes_slice.loc[X, ['ID']] = ['<correct_id>']
# Spooky holloween GY Cube Updated at 6/22/23
poll_cubes_slice.loc[43, ['ID']] = "SpookyBHalloweenGY"

# Replace URL with Cubecobra API Request
poll_cubes_slice['URL'] = 'http://cubecobra.com/cube/api/cubeJSON/' + poll_cubes_slice['ID']
```

```python
# Check that we have things in the correct order
poll_cubes_slice.head(20)
```

```python
# Poll Winners
main_event = poll_cubes_slice[poll_cubes_slice['Poll Winner']==True]
main_event = main_event.reset_index()

# Not Winners, but all Cubes are Winners
side_event = poll_cubes_slice[poll_cubes_slice['Poll Winner']==False]
side_event = side_event.reset_index()

main_event.head(20)
```

```python
# Iterate for the API Pull for Lists

index = 0

df_list = []

for row in main_event.iterrows():
    if rq.get(main_event['URL'][index]).ok == True:
        df = pd.json_normalize(rq.get(main_event['URL'][index]).json().get('cards').get('mainboard'), sep='_')
        df['ID'] = main_event['ID'][index]
        df['Cube Name'] = main_event['Cube Name'][index]
        df['Cubecon Type'] = 'Poll Winner'
        df_list.append(df)
        #print(True)
        index+=1
    else:
        print('Error Encountered At Index:', index)
        index+=1
```

```python
cubes_main_event = pd.concat(df_list)
cubes_main_event.groupby(['Cube Name']).count()
```

```python
# Initial Fields as of 6/3/23:
# Keep Exact same as other Export for Eventual Joining
extract = cubes_main_event.loc[:,[
      'index'
    , 'Cube Name'
    , 'Cubecon Type'
    , 'cardID'
    , 'addedTmsp'
    , 'details_name'
    , 'details_full_name'
    , 'details_artist'
    , 'details_rarity'
    , 'details_color_identity'
    , 'details_colors'
    , 'details_set'
    , 'details_released_at'
    , 'details_cmc'
    , 'details_parsed_cost'
    , 'details_type'
    , 'details_elo'
    , 'details_popularity'
    , 'details_cubeCount'
    , 'details_loyalty'
    , 'details_power'
    , 'details_toughness'
]]

extract.info()
```
